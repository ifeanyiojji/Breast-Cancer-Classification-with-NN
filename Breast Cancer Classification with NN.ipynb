Import the Dependencies

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets 
from sklearn.model_selection import train_test_split
```

Data Collection and Preprocessing 

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
dataframe = pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names)
```

```python
dataframe.head()
```

```python
# adding the target column to the dataframe

dataframe['label'] = breast_cancer_dataset.target
```

Seperating the features and target

```python
x = dataframe.drop(columns='label', axis= 1)
y = dataframe['label']
```

```python
print(x)
```

Splitting the data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Standardized the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)

x_test_std = scaler.transform(x_test)
```

Buidling the Neural Network

```python
#importing tensorflow and keras 

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers


model = keras.Sequential([
                        keras.layers.Flatten(input_shape=(30, )),
                        keras.layers.Dense(20, activation= 'relu'),
                        keras.layers.Dense(2, activation= 'sigmoid')
])
```

```python
# compiling the neural network

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics= ['accuracy'])
```

```python
#training the NN 

history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10, verbose=2)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')


plt.legend(['training data', 'validation data'], loc= 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('loss function')
plt.ylabel('loss')
plt.xlabel('epochs')


plt.legend(['training loss', 'validation loss'], loc= 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```

```python
y_pred = model.predict(x_test_std)
```

```python
print(y_pred.shape)
print(y_pred[0])
```

```python
# converting the prediction probablity to class labels

y_pred_labels = [np.argmax(i) for i in y_pred]
```
